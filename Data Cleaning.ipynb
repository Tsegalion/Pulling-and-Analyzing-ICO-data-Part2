```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Tsega\Scrapy\ico\ico1.csv")
df
```

```python
# Convert start date to a date time
df['Start_date'] = pd.to_datetime(df['Start_date'])

# Convert end date to a date time
df['End_date'] = pd.to_datetime(df['End_date'])
```

```python
# Define the mapping dictionaries for 'Category', 'Token_Role', and 'Platform'
category_mapping = {
    'Blockchain Platform': 'Blockchain Service',
    'Blockchain': 'Blockchain Service',
    'Finance': 'DeFi',
    'Payments': 'DeFi',
    'Market':'Marketplace',
    'Marchket':'Marketplace',
    'Marketing': 'Marketing Service',
    'DEX': 'Exchange',
    'Verification': 'KYC/Security',
    'Advertising': 'Marketing Service',
    'Real Business': 'Business',
    'Social': 'Social Network',
    'Currency': 'DeFi',
    'Artificial Intelligence': 'AI/VR',
    'Predictions': 'Trading Platform',
    'Governance Token': 'DeFi',
    'Mobile': 'Business',
    'Media': 'Business',
    'Mining': 'Business',
    'Crowdfunding': 'Platform',
    'Insurance': 'DeFi',
    'Stablecoin': 'DeFi',
    'Smart Contract':'Platform',
    'SMarcht Contract':'Platform',
    'Wallet': 'DeFi',
    'Dapp': 'Platform',
    'Marchketplace': 'Marketplace',
    'Marchketing': 'Marketing Service',
    'VR': 'AI/VR',
    'KYC': 'KYC/Security',
    'Security': 'KYC/Security',
    'Cloud Storage': 'Cloud',
    'Cloud Computing': 'Cloud',
    'Lending': 'DeFi',
    'Investment': 'Platform',
    'Banking': 'DeFi',
    'Trading': 'Trading Platform',
    'Gambling': 'DeFi',
    'Marketing': 'Marketing Service',
    'Hybrid Intellingence':'AI/VR'
}

token_role_mapping = {
    'Utility Unsold Tokens': 'Utility',
    'Utility\nRegistered Company': 'Utility',
    'South Korea\nRole of Token': 'Utility',
    'Working MVP\nRole of Token': 'Utility',
    'Singapore\nRole of Token': 'Utility',
    '12\nTeam from': 'Utility',
    'USA\nRole of Token': 'Utility',
    '16\nTeam from': 'Utility',
    'Undefined': 'Utility'
}

platform_mapping = {
    'ERC20': 'Ethereum',
    'BEP-20': 'Binance Chain',
    'Flow': 'Flow Network',
    'BEP2': 'Binance Chain',
    'ERC721': 'Ethereum',
    'Matic Blockchain': 'Polygon',
    'ERC777': 'Ethereum',
    'ERC-1404': 'Ethereum',
    'ERC223': 'Ethereum',
    'TRC-10': 'TRON',
    'TRX-20': 'TRON',
    'WRC-20': 'WaykiChain',
    'WAN': 'Wanchain',
    'NEP-5': 'NEO Blockchain',
    'NRC20': 'Nebulas Chain',
    'RST-88': 'Realio Network',
    'OEP-4': 'Ontology Blockchain',
    'Ontology': 'Ontology Blockchain',
    'IRC-20': 'Iost Blockchain',
    'HECO': 'Heco Chain',
    'Own wallet': 'Own Blockchain',
    'Own blockchain': 'Own Blockchain'
}

# Use the 'replace()' method to map the values in the 'Category' column using the 'category_mapping' dictionary
df['Category'] = df['Category'].replace(category_mapping)

# Use the 'replace()' method to map the values in the 'Token_Role' column using the 'token_role_mapping' dictionary
df['Token_Role'] = df['Token_Role'].replace(token_role_mapping)

# Use the 'replace()' method to map the values in the 'Platform' column using the 'platform_mapping' dictionary
df['Platform'] = df['Platform'].replace(platform_mapping)
```

```python
# drop 'NOT SET'
df = df.drop(df[df['Raised_usd'] == 'NOT SET'].index)

# Remove the '$' and ',' and convert to an float
df['Raised_usd'] = df['Raised_usd'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
# Remove USD, $ and the comma from the ICO_price and convert to a float
df['ICO_price'] = df['ICO_price'].str.replace(' USD', '').str.replace('$', '').str.replace(',', '').astype(float)
```

```python
# Remove the '$' and ',' and convert to a float
df['Fundraise_goal'] = df['Fundraise_goal'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
# Renaming the Sold_%
df.rename(columns={'Sold_%' : 'Sold'}, inplace=True)

df['Sold'] = df['Sold'].astype(float)
```

```python
# Remove the ',' and convert to a float
df['Total_token'] = df['Total_token'].str.replace(',', '').astype(float)

df
```
